```python
import pandas as pd
import numpy as np
```

```python
# One way to name the columns
df = pd.read_csv('./auto.csv', header=None)
df.columns = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]

# Another way to do it directly
df = pd.read_csv('./auto.csv', names=["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"])
df.head(5)
```

```python
# So there are some missing values represented by ?
# so we need to replace these

# we used inplace so the changes will be written to the df itself
df.replace('?', np.nan, inplace=True)

# But if we dont then it be done like this
df = df.replace('?', np.nan)
```

```python
# Now we can identify the missing values from the data set using the isnull() and notnull() methods
df.notnull()
df.isnull()
# So the output is a boolean value that shows whether it is a non-null value or not
```

```python
# Counting the missing values
missing_data = df.isnull()
print(type(missing_data))

# for i in missing_data:
#   print(i)

# for i in missing_data.columns:
#   print(i)

# for i in missing_data.columns.values:
#   print(i)


for columns in missing_data:
  # This value_count() method counts the number of occurrence of a unique values in the series
  # so here it count the number how many times TRUE or False appears in the columns
  print((missing_data[columns].value_counts()))
  print()
```

```python
# Getting the names of the columns that have a NAN or there is an empty entry
print(df.columns[df.isnull().any()])

# Here we can replace the normalized-losses, bore, stroke, horsepower and peak-rpm with the mean of their column
# while num-of-doors can be replaced by the most frequent value because most of the cars would have 4 doors
# now for the price since we need to predict it we eliminate this rows that have a null price completely from the data frame
 
# Changing the normalized-losses type to float
print(df['normalized-losses'].dtypes)  # It is object type so it needs to be converted
df['normalized-losses'] = df['normalized-losses'].astype(float)  # converting to float type
print(df['normalized-losses'].dtypes)
```

```python
# GETTING THE IDEA OF ROWS AND COLUMNS CALCULATION
df1 = {'col1': [1, 2, 3], "col2": [10, 20, 30]}
df1 = pd.DataFrame(df1)
print(df1.sum(axis=0))  # So basically along the columns the operation will take place
print(df1.sum(axis=1))  # hero along the row the operation will take place
```

```python
# changing the NAN value to mean
mean = df['normalized-losses'].mean(axis=0)  # calculating the mean of the column
print(mean)

df['normalized-losses'].replace(np.nan, mean, inplace=True)
df['normalized-losses']
```

```python
df['bore'].isnull().value_counts()  # So there are 4 values missing in the bore column

# Checking the datatype of the columhn
df['bore'].dtypes  # object

# converting to float
df['bore'] = df['bore'].astype('float')

# CALCULATING THE MEAN
mean = df['bore'].mean()

# Replacing the data with mean
df['bore'].replace(np.nan, mean, inplace=True)

df['bore'].isnull().value_counts()
```

```python
# 'normalized-losses', 'num-of-doors', 'bore', 'stroke', 'horsepower', 'peak-rpm', 'price'
# Changing the other columns as well

mean = df['stroke'].astype('float').mean()
df['stroke'].replace(np.nan, mean, inplace=True)
df['stroke'].isnull().value_counts()

mean = df['horsepower'].astype('float').mean()
df['horsepower'].replace(np.nan, mean, inplace=True)
df['horsepower'].isnull().value_counts()

mean = df['peak-rpm'].astype('float').mean()
df['peak-rpm'].replace(np.nan, mean, inplace=True)
df['peak-rpm'].isnull().value_counts()
```

```python
# Now for number of doors the values of missing doors we assume is the most common value in the series aso wee assign the empty value to be equal to the value of the most occurring value
num_doors = df['num-of-doors'].value_counts().idxmax() # This gives the most commonly occurring value from the value counts

df['num-of-doors'].replace(np.nan, num_doors, inplace=True)
df['num-of-doors'].isnull().value_counts()
```

```python
# Now its time to drop the price rows which have null values
df.dropna(axis=0, subset=['price'], inplace=True)  # removed the rows with a null price from the data frame

# resetting t5he index because we removed some rows
df.reset_index(drop=True, inplace=True)
df['price'].isnull().value_counts()
```

```python
# Converting the datatypes to correct format
df.dtypes

df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses", 'horsepower']] = df[["normalized-losses", "horsepower"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")

df.dtypes
```

```python
# Now time to standardize, i.e. converting the data into common format 
# like transform mpg to L/100km

df['city-mpg'] = 235 / df['city-mpg']
df['highway-mpg'] = 235 / df['highway-mpg']

df.rename(columns={'city-mpg': 'city-L/100km', 'highway-mpg': 'highway-L/100km'}, inplace=True)

df['city-L/100km']
df['highway-L/100km']

```

```python
# Now time to normalize the data
# So we need to convert the values of the columns to a specific range
# eg: we ne4ed to scale the columns "length", "width" and "height".\
# Target: normalize those variables so their value ranges from 0 to 1

# So here we take the approach of dividing the value of the currentValue / maxValue

df['length']
df['length'] = df['length'] / df['length'].max()
df['length']

# similarly for the width and height
df['width'] = df['width'] / df['width'].max()
df['width']

df['height'] = df['height'] / df['height'].max()
df['height']
```

```python
# Now binning the data
df['horsepower'].value_counts().count() # so here we have 59 unique values

# so we can bin the horsepower into bins of different speeds
df['horsepower'].max()
df['horsepower'].min()

# So the grates and min speeds are 262 to 48
```

```python
import matplotlib.pyplot as plt

plt.hist(df['horsepower'])
plt.xlabel('horsepower')
plt.ylabel('count')
plt.title('horsepower bin')
```

```python
# now we will create three bins LOW, MEDIUM and HIGH of equal space ans sort values on the basis of the bins
num_bins = 3
bins = np.linspace(df['horsepower'].min(), df['horsepower'].max(), num_bins + 1)
bins # this gives us the point or the value at which there are three equally spaced columns
# So here the
# Bin1 starts from 48 to 119.33
# Bin2 119.33 to 190.667
# Bin3 190.66 to 262
```

```python
# Setting the group names
bin_names = ['Low', 'Medium', 'High']

df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=bin_names, include_lowest=True)
df[['horsepower', 'horsepower-binned']]

plt.bar(bin_names, df['horsepower-binned'].value_counts())

plt.xlabel("horsepower")
plt.ylabel("count")
plt.title("horsepower bins")
```

```python
# Normally, you use a histogram to visualize the distribution of bins we created above. 
plt.hist(df['horsepower'], bins=3)

plt.xlabel("horsepower")
plt.ylabel("count")
plt.title("horsepower bins")
```

```python
# Converting the categorical values into usable forms as indicator variables
# we use indicator variables so you can use categorical variables for regression analysis in the later modules.
# eg: the column "fuel-type" has two unique values: "gas" or "diesel". Regression doesn't understand words, only numbers. To use this attribute in regression analysis, you can convert "fuel-type" to indicator variables.

# Creating a dummy variable for the fuel type
dummy_variable_1 = pd.get_dummies(df['fuel-type'])
dummy_variable_1

# renaming the fuel type
dummy_variable_1.rename(columns={'gas': "fuel-type-gas", 'diesel': 'fuel-type-diesel'}, inplace=True)
dummy_variable_1

# adding the dummy columns to the main dataframe 
df = pd.concat([df, dummy_variable_1], axis=1) # axis=1 because we are adding entire columns

# Removing the fuel type column
df.drop('fuel-type', axis=1, inplace=True)
df['fuel-type-gas']
df['fuel-type-diesel']
```

```python
# Creating a dummy for the aspiration now
df['aspiration'].value_counts()

df = pd.concat([df, pd.get_dummies(df['aspiration'])], axis=1)
df.drop('aspiration', axis=1 ,inplace=True)
df
```
